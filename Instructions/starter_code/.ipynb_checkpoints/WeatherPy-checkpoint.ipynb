```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "../output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
#print (cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python

baseurl = "http://api.openweathermap.org/data/2.5/weather?q=" 
my_city_list = []
counter = 0
for city in cities[0:500] :
    counter += 1 
    print ('processing no ' + str(counter) + "  " + city)
    city_json = requests.get(baseurl + city + "&appid=" + weather_api_key)
    #print (reason_json)
    my_city_list.append(city_json.json())

    
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
my_pd = pd.DataFrame(my_city_list)
my_pd
my_pd.to_csv(output_data_file)
print(my_pd)
my_pd
```

```python
# Create Northern and Southern Hemisphere DataFrames
nothern_list = []
southern_list = []

for eachtemp in my_city_list :
    if 'coord' in eachtemp.keys() and 'lat' in eachtemp['coord'] :
        if float(eachtemp['coord']['lat']) >= 0 :
            nothern_list.append(eachtemp)
        else :
            southern_list.append(eachtemp)
            
npd = pd.DataFrame (nothern_list)
spd = pd.DataFrame(southern_list)
npd
spd
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
city_temp_list = []
city_lattitude = []
for eachtemp in nothern_list :
    #print (type(eachtemp.keys()))
    #print(eachtemp)
    if 'main' in eachtemp.keys() and 'temp_max' in eachtemp['main'] :
        city_temp_list.append(eachtemp['main']['temp_max'])
        if 'coord' in eachtemp.keys() and 'lat' in eachtemp['coord'] :
            city_lattitude.append(eachtemp['coord']['lat'])
plt.scatter(city_lattitude,city_temp_list)

x_values = city_lattitude
y_values = city_temp_list
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
print(f"The r-squared is: {rvalue**2}")
```
